### time_zone 매칭 결과 확인 및 수정이 필요한 부분 수정

```python
time_zone = events.loc[:,'country_state':'time_zone'].drop_duplicates()

from IPython.display import HTML
HTML(time_zone.to_html())
```

```python
# time zone data import
time_zone = pd.read_csv("E:/kaggle_Outbrain_Click_Prediction/time_zone_171129.csv")

print(time_zone.head())
print(time_zone.shape)
print(type(time_zone))
```

```python
events = pd.merge(events, time_zone, how='left', on='country_state')  

events.head()
```

```python
events['timestamp_2'] = events['timestamp']
events['timestamp_2'] = events.timestamp_2.dt.tz_localize('America/New_York')

events.head()
```

```python
time_zone_unique = events[['time_zone','timestamp','timestamp_2']]
time_zone_unique = time_zone_unique.drop_duplicates(['time_zone'],keep='last')
time_zone_unique = time_zone_unique.reset_index(drop=True)

time_zone_unique.head()
```

```python
from progressbar import *               # just a simple progress bar
import time, sys

widgets = ['Progress_1st : ', Percentage(), ' ', Bar(marker='=',left='[',right=']'),
           ' ', ETA(), ' ', FileTransferSpeed()] #see docs for other options
pbar = ProgressBar(widgets=widgets, maxval=len(time_zone_unique))                    
pbar.start()

time_difference = []
for i in range(0,len(time_zone_unique)):
    tz = time_zone_unique.loc[i]['time_zone']

    b = time_zone_unique.loc[i]['timestamp_2'].tz_convert(tz) # 현지 적용 시간
    b = b.tz_localize(None)

    c = time_zone_unique.loc[i]['timestamp'] # 뉴욕 기준 시간

    a = b - c # 시차 계산
    if a.days == -1 :
        a = c-b
        time_difference.append(-a)
    else :
        time_difference.append(a)

    del(a,b,c)
    pbar.update(i) #this adds a little symbol at each iteration

pbar.finish()
```

```python
time_difference = pd.DataFrame(time_difference, columns=['time_difference']) 
time_zone_difference = pd.concat([time_zone_unique.reset_index(drop=True), time_difference], axis=1)
time_zone_difference = time_zone_difference[['time_zone','time_difference']]

time_zone_difference.head()
```

```python
events = events.iloc[:,0:11] # time_stamp2 제거
events = pd.merge(events, time_zone_difference, how='left', on='time_zone')  
events['local_timestamp'] = events['timestamp'] + events['time_difference']

events.head(10)
```

```python
pageviews_uuid.geo_location.fillna('', inplace=True) # na 


location = pageviews_uuid['geo_location'].fillna('').str.split('>') # na가 있는 경우 > 를 넣어줌

pageviews_uuid['country'] = location.map(lambda loc: loc[0] if len(loc) > 0 else 'NaN' ) # 첫번째 > 앞 text를 country로 입력 / 없으면 NaN
pageviews_uuid['state'] = location.map(lambda loc: loc[1] if len(loc) > 1 else 'NaN') # 두번째 > 앞 text를 country로 입력 / 없으면 NaN
pageviews_uuid['region'] = location.map(lambda loc: int(loc[2]) if len(loc) > 2 else -1) # 마지막 > 뒤 text를 country로 입력 / 없으면 -1

print(pageviews_uuid.head())
print(pageviews_uuid.shape)
```

```python
country = pageviews_uuid.iloc[:,0:3]

print(country.head())
print(country.shape)
```

```python
country.loc[:,"UserCount"] = country.groupby("geo_location")["geo_location"].transform("count")
country = country.sort_values('UserCount',ascending =[0])

print(country.head())
print(country.shape)
```

```python
country = country.loc[:, ["geo_location", "country", "UserCount"]].drop_duplicates("country", keep="first")
country.sort_values("UserCount", ascending=False, inplace=True)

print(country.head())
print(country.shape)
```
